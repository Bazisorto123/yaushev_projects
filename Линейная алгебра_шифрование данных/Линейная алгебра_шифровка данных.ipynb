## Умножение матриц

В этом задании вы можете записывать формулы в *Jupyter Notebook.*

Чтобы записать формулу внутри текста, окружите её символами доллара \\$; если снаружи —  двойными символами \\$\\$. Эти формулы записываются на языке вёрстки *LaTeX.* 

Для примера мы записали формулы линейной регрессии. Можете их скопировать и отредактировать, чтобы решить задачу.

Работать в *LaTeX* необязательно.

Обозначения:

- $X$ — матрица признаков (нулевой столбец состоит из единиц)

- $y$ — вектор целевого признака

- $P$ — матрица, на которую умножаются признаки

- $w$ — вектор весов линейной регрессии (нулевой элемент равен сдвигу)

Предсказания:

$$
a = Xw
$$

Задача обучения:

$$
w = \arg\min_w MSE(Xw, y)
$$

Формула обучения:

$$
w = (X^T X)^{-1} X^T y
$$

Необходимо ответить на следующий вопрос:
"Признаки умножают на обратимую матрицу. Изменится ли качество линейной регрессии? (Её можно обучить заново.)"


**Ответ:** Не изменится. 

**Обоснование:** 

Формулы для раскрытия скобок:
1. для невыражденных квадратных матрицу
$$
(AB)^{-1} = B^{-1}A^{-1} 
$$
2. 
$$
(AB)^{T} = B^{T}A^{T}
$$
Таким образом, 
$$
a = Xw = X(X^T X)^{-1} X^T y 
$$
Пусть 
$$
Z = XP
$$
, где P - обратимая матрица, тогда


$$
a_1 = XPw_1 = XP((XP)^T XP)^{-1} (XP)^T y = XP(P^{T}X^{T}XP)^{-1}P^{T}X^{T}y
$$

$$
X^{T}XP 
$$

является квадратной матрицей, поэтому можно примернить свойство обратных матриц:

$$
XP(P^{T}X^{T}XP)^{-1}P^{T}X^{T}y = XP(X^{T}XP)^{-1}(P^{T})^{-1}P^{T}X^{T}y
$$

$$
(P^{T})^{-1}P^{T} = E
$$

Поэтому

$$
XP(X^{T}XP)^{-1}(P^{T})^{-1}P^{T}X^{T}y = XP(X^{T}XP)^{-1}X^{T}y = XPP^{-1}(X^{T}X)^{-1}X^{T}y = XPP^{-1}(X^{T}X)^{-1}X^{T}y = X(X^{T}X)^{-1}X^{T}y
$$

$$
a_1 = a_0
$$

Доказано, что результат предсказаний одинаков.

Как связаны параметры линейной регрессии в исходной задаче и в преобразованной?

$$
w_1 = (X^{T}XP)^{-1}(P^{T})^{-1}P^{T}X^{T}y = (X^{T}XP)^{-1}X^{T}y = P^{-1}(X^{T}X)^{-1}X^{T}y = P^{-1}(X^{T}X)^{-1}X^{T}y = P^{-1}w
$$

Таким образом, $$w_1 = P^{-1}w$$

Ответ: параметры $w$ и $w_p$ можно выразить друг через друга так: $$w_1 = P^{-1}w$$

## Алгоритм преобразования

**Алгоритм**

1. Раздеяем данные на признаки и целевые значения.
2. Генерируем случайную квадратную матрицу размером равным количеству признаков.
3. Проверяем эту матрицу на обратимость. (Она должна быть обратимой, ее определитель не равен 0). Если матрица необратима, генерируем новую до тех пор, пока она не пройдет проверку на обратимость.
4. Находим новую матрицу признаков путем умножения исходной на полученную на предыдущем этапе алгоритма.
5. Обучаем модель и находим пресказания с новой матрицей.

**Обоснование**

Качество линейной регрессии не поменяется, так как в пункте 2 было доказано, что перемножение матрицы признаков на обратимую матрицу не влияет на предсказание. При этом обратимая матрица должна иметь размер, равный количеству признаков, иначе мы не сможем выполнить умножение. При этом обратимая матрица каждый раз будет определяться случайным образом, что позволит закодировать исходные данные без влияния на конечный результат.

## Проверка алгоритма

Разделим выборку.

```python
X = data.drop("Страховые выплаты", axis=1)
y = data["Страховые выплаты"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12345)
```

Обучим модель на исходных данных и посчитаем метрику r2.

```python
model = LinearRegression()
model.fit(X_train,y_train)
prediction = model.predict(X_train)
print(r2_score(y_train, prediction))
```

В цикле найдем обратимую матрицу, подходящего для матрицы признаков размера. В цикле же проверим ее на обратимость и будем генерировать случайную матрицу до момента, пока не получим ту, которая удовлетваоряет условию. При этом вероятность получить случайную матрицу с определителем равным 0 крайне мала. Но тем не менее. Далее умножим на эту матрицу исходную, заново обучим модель, уже на этой матрице и снова найдем r2. Если в обоих случаях r2 будут равны, это будет служить подтверждением доказательства из п.2. и верностью алгоритма из п.3

```python
i = 0
while i != 1:
    R = np.random.random(size=(X_train.shape[1],X_train.shape[1]))
    if np.linalg.det(R) != 0:
        A = R
        i = 1
print(A)

X_1 = X_train @ A
model.fit(X_1,y_train)
prediction = model.predict(X_1)
print(r2_score(y_train, prediction))
```

Видно, что r2 равны или максимально близки. Таким образом, задача решается верно.
